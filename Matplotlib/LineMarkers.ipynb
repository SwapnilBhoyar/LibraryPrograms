```python
"""
@Author: Swapnil Bhoyar
@Date: 2021-07-14 
@Last Modified by: Swapnil Bhoyar
@Last Modified time: 2021-08-14
@Title : â€‹ Write a Python program to plot two or more lines and set the line markers.
"""

import Log
import matplotlib.pyplot as plt

def lineMarkers():
    try:
        x = [10,20,30,40,50]
        y = [30,30,30,30,30]
        
        plt.plot(x, y, label = "line 1", marker='o', markerfacecolor='orange', markersize=12)
        plt.plot(y, x, label = "line 2", marker='o', markerfacecolor='green', markersize=12)
        plt.legend()
        plt.show()
    except Exception as e:
        Log.logging.error(e)

lineMarkers()
```

```python
data = pd.read_csv('data.csv')
year = data['Year']
sea_levels = data['CSIRO Adjusted Sea Level']

plt.scatter(year, sea_levels, edgecolors='b')

plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')

plt.title('Rise in Sealevel')
plt.show()
```

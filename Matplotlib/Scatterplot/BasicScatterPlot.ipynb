```python
"""
@Author: Swapnil Bhoyar
@Date: 2021-07-15 
@Last Modified by: Swapnil Bhoyar
@Last Modified time: 2021-08-15
@Title : Matplotlib scatterplot programs.
"""

import matplotlib.pyplot as plt
from pylab import randn
```

```python
X = randn(150)
Y = randn(150)
plt.scatter(X,Y, color='r')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
X = randn(150)
Y = randn(150)
plt.scatter(X,Y, facecolors='none', edgecolors = 'blue')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
